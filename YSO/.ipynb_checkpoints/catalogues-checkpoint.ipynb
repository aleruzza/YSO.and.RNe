## Adding alphas in df4

```python
### Computing alphas for Marton16 ysos

filtersMag = {
#  '2J': [1594.0, 'Jmag', 12350.0] ,
#  '2H': [1024.0, 'Hmag', 16620.0] ,
 '2K': [666.8, 'Ksmag', 21590.0] ,
 'W_3p4': [309.5, 'W1mag', 33526.0] ,
 'W_4p6': [171.8, 'W2mag', 46028.0] ,
 'W_12': [31.7, 'W3mag', 115608.0] ,
 'W_22': [8.36, 'W4mag', 220883.0]}

x = [filtersMag[filt][2] for filt in list(filtersMag)]
wavls = (np.array(x)* u.angstrom ).to(u.micron).value

sedData = []

alphas = []
ealphas = []
xData = np.log10(wavls)
plot_bool = False

for i in range(len(df4)):
    fluxes = []
    efluxes = []
    
    for filt in list(filtersMag):    # Filters with magnitude
        zp = filtersMag[filt][0]* u.jansky
        wave = filtersMag[filt][2]* u.angstrom 
        wave = wave.to(u.micron) # converting angstrom wavelength to micron
        
        magval = df4.iloc[i][str(filtersMag[filt][1])]
        magerr = df4.iloc[i]['e_'+str(filtersMag[filt][1])]
        if np.isnan(magerr):
            magerr = np.max(df4['e'+str(filtersMag[filt][1])]) + 0.1
            
        mag = ufloat(magval,magerr)
        flux = zp*10**(-0.4*mag)

        fluxw = flux*const.c/(wave**2)
        fluxw = fluxw.to(u.erg/u.s/u.cm**2/u.micron)
       
        fluxes += [fluxw.value.n]
        efluxes += [fluxw.value.s]
        
    sedData += [[np.array(fluxes),np.array(efluxes)]]
    
    yData = np.log10(sedData[-1][0]*wavls)
    yError = np.log10(sedData[-1][1]*wavls)
    ind2 = ~np.isnan(yData)    

    try:
        popt,pcov = curve_fit(linear,xData[ind2],yData[ind2],sigma = yError[ind2])
    except Exception as e:
        print(i,'Didnt work',e)
        alphas+= [np.nan]
        ealphas+= [np.nan]
        continue
    
    alphas += [popt[0]]
    ealphas += [np.sqrt(np.diag(pcov))[0]]
    
    if plot_bool:
        plt.errorbar(wavls,sedData[-1][0]*wavls,sedData[-1][1]*wavls,fmt = '.')
        plt.plot(wavls,10**linear(xData,alphas[i],np.log10(sedData[-1][0][0]*wavls[0])),'-')

        ax = plt.gca()
        ax.set_yscale('log')
        ax.set_xscale('log')
        ax.set_xticks(wavls)

        ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
        ax.get_xaxis().set_tick_params(which='minor', size=0)
        ax.get_xaxis().set_tick_params(which='minor', width=0) 

        plt.title(r"Obj no. {}, $\alpha = {:.2f}\pm{:.2f}$".format(i,alphas[i],ealphas[i]))
        plt.ylabel(r'$\lambda F_{\lambda}(ergs/cm^{2}/s)$')
        plt.xlabel(r'$\lambda (\mu m)$')
        plt.show()
        
alphas = np.array(alphas)
ealphas = np.array(ealphas)

### Just for some sanity check for individual seds

i = 100
plt.figure(figsize = (5,3.5))
plt.errorbar(wavls,sedData[i][0]*wavls,sedData[i][1]*wavls,fmt = '.')
plt.plot(wavls,10**linear(xData,alphas[i],np.log10(sedData[i][0][0]*wavls[0])),'-')

ax = plt.gca()
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xticks(wavls)

ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax.get_xaxis().set_tick_params(which='minor', size=0)
ax.get_xaxis().set_tick_params(which='minor', width=0) 

plt.title(r"Obj no. {}, $\alpha = {:.2f}\pm{:.2f}$".format(i,alphas[i],ealphas[i]))
# plt.title(r"$\alpha = {:.2f}\pm{:.2f}$".format(alphas[i],ealphas[i]))
plt.ylabel(r'$\lambda F_{\lambda}(ergs/cm^{2}/s)$')
plt.xlabel(r'$\lambda (\mu m)$')
# plt.savefig("../reports/figures/ExampleSED.png",dpi = 150,bbox_inches = 'tight')
```

```python
### Removing alphas with high errors
ealpha_thresh=0.6

plt.hist(ealphas)
plt.axvline(ealpha_thresh,color = 'r', linestyle = '--')
plt.xlabel(r"$\Delta \alpha$")
plt.ylabel("YSOs")
# plt.savefig("../reports/figures/eAlphas.png",dpi = 100,bbox_inches = 'tight')
plt.show()

good = ealphas < ealpha_thresh
alphas[~good] = np.nan

### Adding alpha values to the dataframe
df4['alpha'] = alphas
df4['ealpha'] = ealphas


### Exploring effects of extiction correction
## Calculating correction term
corr = np.nanmedian(df3.alpha0-df3.alpha)
print("Extinction correction:",corr)

## Visualising correction
plt.plot(df3.alpha,df3.alpha0,'.',color = 'grey',alpha = 0.5)
# plt.plot(ysos_n.alpha[ysos_n.dunham_alpha],ysos_n.alpha0[ysos_n.dunham_alpha],'.',color = 'darkblue')
# plt.plot(df3.alpha0,df3.alpha0,'r-')
plt.plot([-4,4],[-4,4],'-',color = 'darkred',linewidth = 2)
plt.plot([-4,4],[-4+corr,4+corr],'--',color = 'red',linewidth = 2)
plt.ylim(-3,3)
plt.xlim(-3,3)
plt.xlabel(r"Measured $\alpha$ (Dunham+ 15)")
plt.ylabel(r"Corrected $\alpha$' (Dunham+ 15)")
plt.savefig("../reports/figures/ExtinctionCorr.png",dpi = 150,bbox_inches = 'tight')
plt.show()

## Applying constant extinction corr. to computed alpha values
df4.loc[:,'alpha0'] = df4['alpha'] + corr
# df4.loc[pd.isna(df4['alpha0']),'alpha0'] = df4[pd.isna(df4['alpha0'])]['alpha'] + corr
print(sum(pd.isna(df4['alpha0'])))

df4
```

```python
from astroquery.gaia import Gaia
def gaia_query(ra, dec, radius=5 * u.arcsec):
    '''Return an ADQL query string for Gaia DR2 + geometric distances 
        from Bailer-Jones et al. 2018 '''
    query_string = '''SELECT TOP 1 *, DISTANCE(POINT('ICRS',g.ra, g.dec), 
                    POINT('ICRS', %s, %s)) as r
                    FROM gaiadr2.gaia_source AS g, external.gaiadr2_geometric_distance AS d
                    WHERE g.source_id = d.source_id AND CONTAINS(POINT('ICRS',g.ra, g.dec), 
                    CIRCLE('ICRS',%15.10f, %15.10f,%15.10f))=1 ORDER BY r ASC''' % \
        (ra, dec, ra, dec, radius.to(u.degree).value) 
    return(query_string)

def gaia_query_edr3(ra, dec, radius=5 * u.arcsec):
    '''Return an ADQL query string for Gaia DR2 + geometric distances 
        from Bailer-Jones et al. 2018 '''
    query_string = '''SELECT TOP 1 *, DISTANCE(POINT('ICRS',g.ra, g.dec), 
                    POINT('ICRS', %s, %s)) as r
                    FROM gaiaedr3.gaia_source AS g
                    WHERE CONTAINS(POINT('ICRS',g.ra, g.dec), 
                    CIRCLE('ICRS',%15.10f, %15.10f,%15.10f))=1 ORDER BY r ASC''' % \
        (ra, dec, ra, dec, radius.to(u.degree).value) 
    return(query_string)

def gaia_query_dr3(ra, dec, radius=5 * u.arcsec):
    '''Return an ADQL query string for Gaia DR2 + geometric distances 
        from Bailer-Jones et al. 2018 '''
    query_string = '''SELECT TOP 1 *, DISTANCE(POINT('ICRS',g.ra, g.dec), 
                    POINT('ICRS', %s, %s)) as r
                    FROM gaiadr3.gaia_source AS g
                    WHERE CONTAINS(POINT('ICRS',g.ra, g.dec), 
                    CIRCLE('ICRS',%15.10f, %15.10f,%15.10f))=1 ORDER BY r ASC''' % \
        (ra, dec, ra, dec, radius.to(u.degree).value) 
    return(query_string)

## Test
gaia_cols = ['parallax','parallax_error']
i=7
query_string = gaia_query_dr3(ysos_n.ra.iloc[i],ysos_n.dec.iloc[i])
job = Gaia.launch_job(query_string, verbose=False)
a = job.get_results()
a[gaia_cols].to_pandas().iloc[0]
```
